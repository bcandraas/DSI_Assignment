## Soal 1
1 TABEL TOTAL REKAPITULASI PENJUALAN

```python
import pandas as pd

# Load files
dokter = pd.read_excel("Data Jasa Medis Detail - Dokter (2022-08 Agustus 1-31).xlsx")
perawat = pd.read_excel("Data Jasa Medis Detail - Perawat (2022-08 Agustus 1-31).xlsx")
bc = pd.read_excel("Data Jasa Medis Detail - BC (2022-08 Agustus 1-31).xlsx")
```

```python
# Check table Dokter
print("Dokter:", dokter.columns)
```

```python
# Check table Perawat
print("Perawat:", perawat.columns)
```

```python
#Check table BC
print("BC:", bc.columns)
```

```python
# Merge Dokter - Perawat into new table
df_merge = pd.merge(dokter, perawat,
                    on=['KD PASIEN','NAMA PASIEN','TANGGAL','NAMA PRODUK','HARGA/UNIT','QTY'],
                    how='outer')

# Merge new table - BC
df_merge = pd.merge(df_merge, bc,
                    on=['KD PASIEN','NAMA PASIEN','TANGGAL','NAMA PRODUK','HARGA/UNIT','QTY'],
                    how='outer')

df_merge
```

```python
# Remove Duplicates
df_merge = df_merge.drop_duplicates()
df_merge.duplicated().sum()
```

```python
# Select the columns to be used
df_merge = df_merge[['KD PASIEN','NAMA PASIEN','TANGGAL','NAMA DOKTER','NAMA PERAWAT',
                     'NAMA BC','NAMA PRODUK','HARGA/UNIT','QTY',
                     'FEE RP DOKTER','FEE RP PERAWAT','FEE RP BC']]

df_merge
```

```python
# Check missing value (NaN)
df_merge.isna().sum()
```

```python
# Change NaN to 0 in FEE DOKTER, PERAWAT and BC
df_merge['FEE RP DOKTER'] = df_merge['FEE RP DOKTER'].fillna(0)
df_merge['FEE RP PERAWAT'] = df_merge['FEE RP PERAWAT'].fillna(0)
df_merge['FEE RP BC'] = df_merge['FEE RP BC'].fillna(0)

df_merge
```

```python
# Check missing value (NaN) for FEE RP DOKTER, PERAWAT and BC
df_merge.isna().sum()
```

```python
# Rename columns name
df_merge = df_merge.rename(columns={
    "FEE RP DOKTER": "FEE DOKTER",
    "FEE RP PERAWAT": "FEE PERAWAT",
    "FEE RP BC": "FEE BC",})

df_merge
```

```python
# Export to excel file
df_merge.to_excel("Rekap_Soal1.xlsx", index=False)
```

## Soal 2
3 TABEL (TABEL DOKTER, TABEL PERAWAT, TABEL BC)

```python
import pandas as pd

# Load files
dokter = pd.read_excel("Data Jasa Medis Detail - Dokter (2022-08 Agustus 1-31).xlsx")
perawat = pd.read_excel("Data Jasa Medis Detail - Perawat (2022-08 Agustus 1-31).xlsx")
bc = pd.read_excel("Data Jasa Medis Detail - BC (2022-08 Agustus 1-31).xlsx")
```

```python
# Check Missing Value
dokter.isna().sum()
```

```python
# Check Missing Value
perawat.isna().sum()
```

```python
# Check Missing Value
bc.isna().sum()
```

```python
# Count total Group By Nama Dokter
df_dokter = dokter.groupby("NAMA DOKTER").agg(
    Total_Omset=("TOTAL", "sum"),
    Total_Fee=("FEE RP DOKTER", "sum")
).reset_index()

# Export to excel file
df_dokter.to_excel("RekapDokter_Soal2.xlsx", index=False)

df_dokter
```

```python
# Count total Group By Nama Perawat
df_perawat = perawat.groupby("NAMA PERAWAT").agg(
    Total_Omset=("TOTAL", "sum"),
    Total_Fee=("FEE RP PERAWAT", "sum")
).reset_index()

# Export to excel file
df_perawat.to_excel("RekapPerawat_Soal2.xlsx", index=False)

df_perawat
```

```python
# Count total Group By Nama BC
df_bc = bc.groupby("NAMA BC").agg(
    Total_Omset=("TOTAL", "sum"),
    Total_Fee=("FEE RP BC", "sum")
).reset_index()

# Export to excel file
df_bc.to_excel("RekapBC_Soal2.xlsx", index=False)

df_bc
```

## Soal 3
1 TABEL REKAPITLASI TREATMENT

Tabel yang digunakan adalah tabel yang sudah dibuat di Soal 1

```python
# Show table from Soal 1
df_merge
```

```python
# Create new table Group By Nama Produk
df_treatment = df_merge.groupby("NAMA PRODUK").agg({
    "QTY": "sum",
    "HARGA/UNIT": "first",
    "FEE DOKTER": "sum",
    "FEE PERAWAT": "sum",
    "FEE BC": "sum"
}).reset_index()

df_treatment
```

```python
# Count Total Omset
df_treatment["TOTAL OMSET"] = df_treatment["QTY"] * df_treatment["HARGA/UNIT"]

df_treatment
```

```python
# Sort from the biggest turnover value
df_treatment = df_treatment.sort_values(by="TOTAL OMSET", ascending=False)

df_treatment
```

```python
# Export to excel file
df_treatment.to_excel("RekapTreatment_Soal3.xlsx", index=False)
```
